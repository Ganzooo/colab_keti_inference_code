## 💻 Open3D 결과
demo/pcd_demo.py를 실행하면 Open3D 시각화 창이 나타납니다. 이 창에서는:

- x, y, z 키: 각 축 평면으로 뷰 정렬
- h 키: 단축키 도움말 표시
- 창을 닫으려면 Esc 키 또는 우측 상단 닫기 버튼 클릭
